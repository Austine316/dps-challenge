```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df2 = pd.read_csv('data.csv')
df2.head(10)
```

```python
dff = df2[['MONATSZAHL', 'AUSPRAEGUNG', 'JAHR', 'MONAT', 'WERT']]
dff = dff.loc[(dff['MONAT'] != 'Summe')]
dff.head(14)
```

```python
dff.dropna(subset=['WERT'], axis=0, inplace=True)
dff.head(5)
```

```python
# df['MONAT'] = month
dff['MONAT'] =  pd.to_datetime(dff['MONAT'], format = '%Y%m')
dff
```

```python
for ii in range(len(dff['MONATSZAHL'].unique())):
    cat = dff['MONATSZAHL'].unique()[ii]
    print(cat)
    dfff = dff.copy()
    dfff = dfff.loc[(dfff['MONATSZAHL'] == cat)]
    dfff = dfff[['MONAT', 'WERT']]
    
    dfff.plot(x='MONAT', figsize=(15,7))
    plt.show()
```

```python
df = dff.copy()
df = df.loc[(dff['MONAT'] != 'Summe') & (df['MONATSZAHL']=='Alkoholunfälle') & (df['AUSPRAEGUNG']=='insgesamt')]
df = df[['MONAT', 'WERT']]
df.to_csv('dat2.csv', index = False)
```

```python
df = pd.read_csv('dat2.csv',index_col='MONAT',parse_dates=True)
df.plot(figsize=(15,10))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
print(df['WERT'].isna().sum())
df = df.dropna(how='any',axis=0) 
print(df['WERT'].isna().sum())
```

```python
df = df.iloc[::-1]
df.head(4)
```

```python
train = df.iloc[:240]
test  = df.iloc[240:]
print(len(train), len(test))
```

```python
train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df.head(),df.tail()
```

```python
scaler.fit(train)
scaled_train = scaler.transform(train)
scaled_test  = scaler.transform(test)
```

```python
scaled_train
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# define generator
n_input    = 3
n_features = 1
generator  = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
X,y = generator[1]
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n {y}')
```

```python
X.shape
```

```python
# We do the same thing, but now instead for 12 months
n_input   = 12
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import LSTM
```

```python
# define model
model = Sequential()
model.add(LSTM(100, activation='relu',return_sequences=False, input_shape=(n_input, n_features)))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
# fit model
model.fit(generator,epochs=100)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
last_train_batch = scaled_train[-n_input:]
```

```python
last_train_batch = last_train_batch.reshape((1, n_input, n_features))
```

```python
model.predict(last_train_batch)
```

```python
scaled_test[0]
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(test)):
    
    # get the prediction value for the first batch
    current_pred = model.predict(current_batch)[0]
    print(current_pred)
    
    # append the prediction into the array
    test_predictions.append(current_pred) 
    
    # use the prediction to update the batch and remove the first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
year  = []
month = []
for ii in test.index:
    year.append(str(ii)[:4])
    month.append(str(ii)[5:7])
```

```python
test["year"]  = year
test["month"] = month
test.head()
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```

```python
test['Predictions'] = true_predictions
```

```python
test
```

```python
test.plot(figsize=(14,5))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(test['WERT'],test['Predictions']))
print(rmse)
```

```python
test.to_csv('prediction.txt', index = True)
```
